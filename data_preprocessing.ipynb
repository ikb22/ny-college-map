```python
import pandas as pd
```

```python
# First dataset containing all 4 year colleges in NY
df = pd.read_csv('ny_four_year_colleges.csv')
df = df.drop(columns=["City"])
# Remove special characters to cast to int/float
df = df.replace(',', '', regex=True)
df = df.replace('\$', '', regex=True)
df[["In-State Cost", "Out-State Cost"]] = df[["In-State Cost", "Out-State Cost"]].replace('-', '0', regex=True)
df[["Enroll", "In-State Cost", "Out-State Cost"]] = df[["Enroll", "In-State Cost", "Out-State Cost"]].apply(pd.to_numeric)
df.info(verbose=True)
df.head()
```

```python
# Second dataset containing addressses of colleges in NY
df2 = pd.read_csv('ny_colleges_addresses.csv')
df2 = df2.drop(columns=["NUMBER OF UG STUDENTS", "FOUNDED", "IN-STATE TUITION", "OUT OF STATE TUITION"])
df2 = df2.rename(columns={"NAME": "School", "ADDRESS": "Address", "TYPE": "Type", "ACCEPTANCE RATE": "Acceptance Rate"})
df2.info(verbose=True)
df2.head()
```

```python
common = df.merge(df2, on=["School"])
df2.info(verbose=True)
common.head()
```

```python
# Third dataset containing statistics on colleges in the US
df3 = pd.read_csv("us_college_statistics.csv")
df3 = df3.drop(columns=["ACCEPTANCE RATE", "YIELD", "% SUBMITTING SAT", "% SUBMITTING ACT", "SUPERSCORING", "SCORE CHOICE", "TEST OPTIONAL/BLIND", "SELF-REPORTING SAT/ACT SCORE"])
df3 = df3.rename(columns={"SCHOOL": "School", "SAT TOTAL 25th-75th%": "SAT Total 25th-75th%", "ACT TOTAL 25th-75th%": "ACT Total 25th-75th%"})
df3.info(verbose=True)
df3.head()
```
